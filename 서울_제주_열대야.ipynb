```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import font_manager, rc

# Use Nanum Gothic font
font_path = 'C:/Users/YTN/Downloads/Nanum_Gothic/NanumGothic-Regular.ttf'  # Full path to Nanum Gothic Regular font
font_manager.fontManager.addfont(font_path)  # Register the font
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)  # Set the font for use in the plot

# Load the data
file_path = '열대아일수/서울_열대야.xlsx'
data = pd.read_excel(file_path)

# Filter the data for 서울
seoul_data = data[data['지점'] == '서울']

# Drop rows with missing start or end dates
seoul_data = seoul_data.dropna(subset=['시작일', '종료일'])

# Convert 시작일 and 종료일 to datetime format
seoul_data['시작일'] = pd.to_datetime(seoul_data['시작일'])
seoul_data['종료일'] = pd.to_datetime(seoul_data['종료일'])

# Extract the year for the y-axis and calculate the difference between 종료일 and 시작일
seoul_data['연도'] = seoul_data['시작일'].dt.year
seoul_data['열대야 기간'] = (seoul_data['종료일'] - seoul_data['시작일']).dt.days

# Create the plot
plt.figure(figsize=(10, 8))
for index, row in seoul_data.iterrows():
    plt.barh(row['연도'], row['열대야 기간'], left=row['시작일'].dayofyear, color='skyblue')

# Format the x-axis to show the days in the year (Jan 1 to Dec 31)
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))

# Adding labels and title in Korean
plt.title('서울의 연도별 열대야 기간', fontsize=14)
plt.xlabel('기간 (월/일)', fontsize=12)
plt.ylabel('연도', fontsize=12)
plt.grid(True)

# Save the plot image to the desktop
plt.savefig('C:/Users/YTN/Desktop/서울_열대야_기간_그래프.png', dpi=300)  # Save the plot as a PNG file

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import font_manager, rc

# Use Nanum Gothic font
font_path = 'C:/Users/YTN/Downloads/Nanum_Gothic/NanumGothic-Regular.ttf'  # Full path to Nanum Gothic Regular font
font_manager.fontManager.addfont(font_path)  # Register the font
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)  # Set the font for use in the plot

# Load the data
file_path = '열대아일수/서울_열대야.xlsx'
data = pd.read_excel(file_path)

# Filter the data for 서울
seoul_data = data[data['지점'] == '서울']

# Drop rows with missing start or end dates
seoul_data = seoul_data.dropna(subset=['시작일', '종료일'])

# Convert 시작일 and 종료일 to datetime format
seoul_data['시작일'] = pd.to_datetime(seoul_data['시작일'])
seoul_data['종료일'] = pd.to_datetime(seoul_data['종료일'])

# Extract the year for the y-axis and calculate the difference between 종료일 and 시작일
seoul_data['연도'] = seoul_data['시작일'].dt.year
seoul_data['열대야 기간'] = (seoul_data['종료일'] - seoul_data['시작일']).dt.days

# Define the current year to highlight (e.g., 2024)
current_year = 2024
current_year_data = seoul_data[seoul_data['연도'] == current_year]

# Create the plot
plt.figure(figsize=(12, 8))

# Loop through each row and plot the consecutive periods of 열대야
for index, row in seoul_data.iterrows():
    # The bar represents the period between 시작일 and 종료일
    start_day = row['시작일'].dayofyear
    duration = row['열대야 기간']
    color = 'red' if row['연도'] == current_year else 'skyblue'  # Highlight 2024 in red
    plt.barh(row['연도'], duration, left=start_day, color=color)

    # Add a text label for 2024
    if row['연도'] == current_year:
        plt.text(start_day + duration + 1, row['연도'], f'{duration}일',
                 va='center', ha='left', color='red', fontsize=10)

# Format the x-axis to show the days in the year with Korean month labels
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m월 %d일'))  # Format in Korean (월/일)

# Adding labels and title in Korean
plt.title('서울의 연도별 연속 열대야 기간 (2024년 강조)', fontsize=14)
plt.xlabel('기간 (월/일)', fontsize=12)
plt.ylabel('연도', fontsize=12)
plt.grid(True, axis='x')

# Save the plot image to the desktop
plt.savefig('C:/Users/YTN/Desktop/서울_연속_열대야_기간_그래프_2024_강조_한글축.png', dpi=300)  # Save the plot as a PNG file

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load the data
file_path = '열대야일수/서울_시작_마지막.xlsx'
seoul_data = pd.read_excel(file_path)

# Convert 가장 빠른 날짜 and 가장 늦은 날짜 to datetime format
seoul_data['가장 빠른 날짜'] = pd.to_datetime(seoul_data['가장 빠른 날짜'])
seoul_data['가장 늦은 날짜'] = pd.to_datetime(seoul_data['가장 늦은 날짜'])

# Extract only the month and day from '가장 늦은 날짜' and create a new column 'month_day'
seoul_data['month_day'] = seoul_data['가장 늦은 날짜'].dt.month * 100 + seoul_data['가장 늦은 날짜'].dt.day

# Sort the dataframe based on 'month_day' to get the latest dates
seoul_data_sorted = seoul_data.sort_values(by='month_day', ascending=False)

# Select the top 3 latest dates
top_3_latest = seoul_data_sorted.head(3)

# Create the plot for 서울
plt.figure(figsize=(12, 8))

# Plot the bars starting at 가장 빠른 날짜 and ending at 가장 늦은 날짜
for index, row in seoul_data.iterrows():
    # The bar represents the period between 가장 빠른 날짜 and 가장 늦은 날짜
    start_day = row['가장 빠른 날짜'].dayofyear
    duration = (row['가장 늦은 날짜'] - row['가장 빠른 날짜']).days
    color = 'red' if row['연도'] in top_3_latest['연도'].values else 'skyblue'  # Highlight top 3 in red
    plt.barh(row['연도'], duration, left=start_day, color=color)

    # Add text labels for the top 3 years
    if row['연도'] in top_3_latest['연도'].values:
        plt.text(start_day + duration + 2, row['연도'], f'{row["연도"]}', va='center', ha='left', color='black', fontsize=10)

# Format the x-axis to show the days in the year (Jan 1 to Dec 31) in Korean format
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m월 %d일'))  # Format in Korean (월/일)

# Adding labels and title in Korean
plt.title('서울의 연도별 여름기간', fontsize=14)
plt.xlabel('기간 (월/일)', fontsize=12)
plt.ylabel('연도', fontsize=12)
plt.grid(True, axis='x')
plt.savefig('C:/Users/YTN/Desktop/서울의 연도별 여름기간.png', dpi=300)  # Save the plot as a PNG file

# Display the plot
plt.show()
```
